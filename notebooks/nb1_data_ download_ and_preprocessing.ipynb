# Data download and Preprocessing notebook

The purpose of this notebook is to download and preprocess the data that will be used for VHVM training and portfolio analysis.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import os
```

```python
#creating my tickers, choice of etfs are explained in the markdown cell above
tickers = [
    "SPY","QQQ","IWM","XLF","XLK",
    "EFA","EEM",
    "SHY","IEF","TLT","LQD",
    "GLD","DBC","USO",
    "VNQ"
]
```

```python
#data download
data = yf.download(tickers, start="2010-01-01", end="2025-01-01", auto_adjust=True)
```

```python
#checking the multiindex dataframe 
data.head()
```

```python
#taking a look at the multi-index dataframe structure column-wise,i only want the close columns to calculate returns later
print(data.columns)
```

```python
#collecting close prices and checking how many NaN values exist in the data frame
close = data["Close"]
print(close.head())
print(close.isna().sum())
```

## Computing Log-returns

Log returns are computed and normalised in the next cell. Log returns express percentage changes in continuous terms, which standardises returns across assets and allows easier comparison.

```python
#computing log returns and normalising data to mean 0 and variance 1.
returns = np.log(close / close.shift(1)).dropna()

returns = (returns - returns.mean()) / returns.std()
```

```python
#checking the data is now normalised 
print(returns.mean())
print((returns.std())**2)
```

```python
#creating train,validation and test splits
n = len(returns)
train_end = int(0.6*n)
val_end   = int(0.8*n)

train = returns.iloc[:train_end]
val   = returns.iloc[train_end:val_end]
test  = returns.iloc[val_end:]
```

```python
#saving data to the /data folder
os.makedirs("../data/processed", exist_ok=True)
train.to_csv("../data/processed/returns_train.csv")
val.to_csv("../data/processed/returns_val.csv")
test.to_csv("../data/processed/returns_test.csv")
close.to_csv("../data/processed/prices_full.csv")
```
